```python
def rect_aim(img,initx,inity,width,height):
    startx = rand_ints(initx,width)
    starty =  rand_ints(inity,height)
    cv2.rectangle(img,(startx,starty),(startx+width//20,starty+width//20),(255,0,0),2) #The Impact Rectangle 

    return img
    
    
    
```
